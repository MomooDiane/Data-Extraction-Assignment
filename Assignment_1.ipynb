Extraction of the tables from the database

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("hopital.db")
cursor = conn.cursor()
create_tables_sql = """
CREATE TABLE Employees (
    medicare_card_num VARCHAR(255) PRIMARY KEY,
    first_name VARCHAR(255) NOT NULL,
    last_name VARCHAR(255) NOT NULL,
    date_of_birth DATE NOT NULL,
    phone_number VARCHAR(255),
    address VARCHAR(255),
    city VARCHAR(255),
    province VARCHAR(255),
    postal_code VARCHAR(10),
    citizenship VARCHAR(255),
    email_address VARCHAR(255) UNIQUE
);

CREATE TABLE Facilities (
    name VARCHAR(255) NOT NULL,
    web_address VARCHAR(255) NOT NULL,
    capacity INT,
    type VARCHAR(255),
    address VARCHAR(255),
    city VARCHAR(255),
    province VARCHAR(255),
    postal_code VARCHAR(10),
    phone_number VARCHAR(20),
    CONSTRAINT PK_Facilities PRIMARY KEY (name, web_address)
);

CREATE TABLE Infections (
    type VARCHAR(255) PRIMARY KEY
);

CREATE TABLE Vaccines (
    type VARCHAR(255) PRIMARY KEY
);

CREATE TABLE Infected (
    medicare_card_num VARCHAR(12),
    type VARCHAR(255),
    infection_count INT,
    date DATE,
    PRIMARY KEY (medicare_card_num, type, infection_count),
    FOREIGN KEY (medicare_card_num) REFERENCES Employees(medicare_card_num),
    FOREIGN KEY (type) REFERENCES Infections(type)
);

CREATE TABLE Has_Received (
    medicare_card_num VARCHAR(12),
    vaccine_type VARCHAR(255),
    facility_name VARCHAR(255),
    web_address VARCHAR(255),
    date DATE,
    dose_num INT,
    PRIMARY KEY (facility_name, web_address, medicare_card_num, vaccine_type, dose_num),
    FOREIGN KEY (vaccine_type) REFERENCES Vaccines(type),
    FOREIGN KEY (facility_name, web_address) REFERENCES Facilities(name, web_address),
    FOREIGN KEY (medicare_card_num) REFERENCES Employees(medicare_card_num)
);

CREATE TABLE Administers (
    facility_name VARCHAR(255),
    web_address VARCHAR(255),
    vaccine_type VARCHAR(255),
    PRIMARY KEY (facility_name, web_address, vaccine_type),
    FOREIGN KEY (facility_name, web_address) REFERENCES Facilities(name, web_address),
    FOREIGN KEY (vaccine_type) REFERENCES Vaccines(type)
);

CREATE TABLE Managed (
    facility_name VARCHAR(255),
    web_address VARCHAR(255),
    medicare_card_num VARCHAR(12),
    start_date DATE,
    end_date DATE,
    PRIMARY KEY (medicare_card_num, start_date),
    FOREIGN KEY (facility_name, web_address) REFERENCES Facilities(name, web_address),
    FOREIGN KEY (medicare_card_num) REFERENCES Employees(medicare_card_num)
);


CREATE TABLE Worked (
    facility_name VARCHAR(255),
    web_address VARCHAR(255),
    medicare_card_num VARCHAR(12),
    start_date DATE,
    end_date DATE,
    role VARCHAR(255),
    PRIMARY KEY (medicare_card_num, start_date),
    FOREIGN KEY (facility_name, web_address) REFERENCES Facilities(name, web_address),
    FOREIGN KEY (medicare_card_num) REFERENCES Employees(medicare_card_num)
 );

"""
```

```python
cursor.executescript(create_tables_sql)
conn.commit()
```

Query 1: Query that retrieves a list of all employees, along with the details of their vaccinations if they have been vaccinated. If an employee hasn't been vaccinated, the ‘vaccine_name’ and ‘facility_name’ fields will be NULL for that employee (outer-join)

```python
# Merging employees and hasReceived on medicare_card_num
merged_df = employees_df.merge(hasReceived_df, on='medicare_card_num', how='left')

# Merging the result with facilities_df on web_address and name (renaming it to facility_name)
final_df = merged_df.merge(facilities_df, left_on=['facility_name', 'web_address'], right_on=['name', 'web_address'], how='left')

print(final_df[['medicare_card_num', 'first_name', 'last_name', 'vaccine_type', 'facility_name']])

```

Query 3: Give the first and last names of Employees that are actively working (eg. Has no end_date)
